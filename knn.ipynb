```python
import numpy as np

def calc_dist(point1, point2):
    return np.linalg.norm(point1 - point2)

def majority_num(labels):
    tmp = {}

    for item in labels: 
        if str(item) not in tmp.keys(): 
            tmp[str(item)] = 1
        else: 
            tmp[str(item)] = tmp[str(item)] + 1
    max_value = 0
    returned_label = -1

    for key in tmp: 
        value = tmp[key]
        if value >= max_value: 
            max_value = value
            returned_label = key
    return int(returned_label)




def KNNClassifier(x_train, y, x_labeled, n_neighbors): 
    # empty list for predictions
    predictions = []

    # for each point in the x dataset
    for point in x_labeled:
        
        # initialize an empty array to store distances to other points
        distances = np.zeros(len(x_train))

        # for each point in the training set
        for i in range(len(x_train)):
            # calculate the distance to that point 
            distances[i] = calc_dist(x_train[i,:], point)

        # sort ascending distance but return the index instead of distance value
        # and save the first N points
        idx_closest_points = np.argsort(distances)[:n_neighbors]

        # save the labels 
        labels = y[idx_closest_points]

        # find the majority number that appears in the labels
        pred = majority_num(labels)

        # append prediction to predictions
        predictions.append(pred)
    return predictions

```

```python
from sklearn.metrics import accuracy_score
from sklearn.datasets import load_iris
from numpy.random import randint
 
#Loading the Data
iris= load_iris()
 
# Store features matrix in X
X= iris.data
#Store target vector in 
y= iris.target
 
 
#Creating the training Data
train_idx = xxx = randint(0,150,100)
X_train = X[train_idx]
y_train = y[train_idx]
 
#Creating the testing Data
test_idx = xxx = randint(0,150,50) #taking 50 random samples
X_test = X[test_idx]
y_test = y[test_idx]
 
#Applying our function 
y_pred = KNNClassifier(X_train,y_train,X_test , 7)
 
#Checking the accuracy
accuracy_score(y_test, y_pred)
```
