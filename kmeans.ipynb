```python
import random 
import numpy as np
def kmeans(numClusters, X):

    # init the centroids
    centroids =  np.asarray(random.choices(X, k=3))


    # init the cluster lables 
    y = np.zeros(len(X))

    diff = 1
    # while the difference between the previous centroids and new centroids is greater than threshold
    while (diff > 0.001):
        # loop through data 
        idx = 0
        # for each element in X
        for element in X: 

            # find nearest centroid
            closest = 10000000000
            for cent in centroids: 
                dist = np.linalg.norm(element - cent)
                if dist < closest: 
                    closest = dist
                    cent_idx = np.where(centroids == cent)[0][0]

            # assign to that centroid        
            y[idx] = cent_idx
            idx = idx + 1

        # calculate the new centroid 
        new_cent_list = []
        # For each cluster
        for i in range(len(centroids)):
            # start sum at 0s
            sum = np.zeros(len(centroids[0]))

            # start counter at 0
            counter = 0

            # for each element in Y
            for j in range(len(y)):
                # if the value of Y (the cluster index) is equal to the cluster that we are looking at now 
                if y[j] == i: 
                    # add to the sum
                    tmp = np.asarray(X[j])

                    # add
                    sum = np.add(tmp, sum)

                    # increment the counter
                    counter = counter + 1

            # add the new centroid to an array 
            new_cent_list.append(np.asarray(np.divide(sum, counter)))

        new_cent_list = np.asarray(new_cent_list)
        # calculate mean difference of centroids 

        
        
        diff = 0 
        # for each new centroid, calculate the distance to the old centroid and add
        for i in range(len(centroids)): 
            diff = diff + np.linalg.norm(centroids[i] - new_cent_list[i])

        # divide by number of centroids to get the average distance moved
        diff /= len(centroids)

        # set the old centroid list to the new centroids list 
        centroids = new_cent_list
        #repeaet 
        
    # return labels and centroids
    return y, centroids
```
